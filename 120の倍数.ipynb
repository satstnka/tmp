## 命題

$n^{62}+15n^{27}+15n^{32}+n^{31}+20n^{22}+24n^{14}+20n^{11}+24n^{7}$ は120 で割り切れる

## 検証

```python
import sympy

n = sympy.Symbol('n')
a = sympy.Symbol('a')
b = sympy.Symbol('b')

f = (n ** 62) + 15 * (n ** 27) + 15 * (n ** 32) + (n ** 31) + 20 * (n ** 22) + 24 * (n ** 14) + 20 * (n ** 11) + 24 * (n ** 7)
f
```

```python
for i in range(1, 11):
    print(f'f({i})/120 = {f.subs(n, i) / 120}')
```

$n=1 \sim 10$ のとき120で割り切れるので命題は成り立ちそう。

## 証明

(与式) $= f(n)$ とする

120 を因数分解すると

```python
sympy.factorint(120)
```

$2^{3}\times 3\times 5$

よって以下の3つの条件を満たせば良い
   1. 与式は　$2^{3}$ で割り切れる
   2. 与式は 3 で割り切れる
   3. 与式は 5 で割り切れる

### (1) $2^{3}$ で割り切れることの証明

i. $n$ が偶数の時

$n = 2a$ とおける。 ($a$ は自然数)

$f(n)$ の $n$ に $2a$ を代入して

```python
expr1 = f.subs(n, 2 * a)
expr1
```

これを因数分解して

```python
sympy.factor(expr1)
```

これは　1024 $(2^{10})$ の倍数なので $2^{3}$ で割り切れる

ii. $n$ が奇数の時

$n = 2a - 1$ ($a$ は自然数) とおける

$f(n)$ の $n$ に $2a - 1$ を代入して

```python
expr2 = f.subs(n, 2 * a - 1)
expr2
```

これを因数分解して

```python
expr3 = sympy.factor(expr2)
expr3
```

$a$ が偶数のときは $a=2b$ ($b$ は自然数) とおけるので、この式は

```python
expr4 = expr3.subs(a, 2 * b)
sympy.factor(expr4)
```

となり、これは8の倍数となる。

$a$ が奇数のときは $a = 2b - 1$ ($b$ は自然数) とおけるので、この式は

```python
expr5 = expr3.subs(a, 2 * b - 1)
sympy.factor(expr5)
```

となり、こちらも８の倍数となる。

よって与式は８の倍数

### (2) ３で割り切れることの証明

- $n$ が ３で割り切れる数の場合　$n=3a$
- $n$ が 3で割って1余る数の場合 $n=3a-2$
- $n$ が 3で割って2余る数の場合 $n=3a-1$

とおける。($a$ は自然数)

```python
array = [3 * a, 3 * a - 2, 3 * a - 1]
for element in array:
    print(f'n={element}のとき')
    expr = f.subs(n, element)
    factor = sympy.factor(expr)
    print(f'    {factor}')
```

これらはいずれも３の倍数である。

よって。与式は3の倍数である。

### (3) 5で割り切れることの証明

- $n$ が 5で割り切れる数の場合　$n=5a$
- $n$ が 5で割って1余る数の場合 $n=5a-4$
- $n$ が 5で割って2余る数の場合 $n=5a-3$
- $n$ が 5で割って3余る数の場合 $n=5a-2$
- $n$ が 5で割って4余る数の場合 $n=5a-1$

とおける。($a$ は自然数)

```python
array = [5 * a, 5 * a - 4, 5 * a - 3, 5 * a - 2, 5 * a - 1]
for element in array:
    print(f'n={element}のとき')
    expr = f.subs(n, element)
    factor = sympy.factor(expr)
    print(f'    {factor}')
```

これらはいずれも５の倍数である。

(1) (2) (3) より、与式は 8 と 3 と 5 の倍数なので、１２０で割り切れる。
